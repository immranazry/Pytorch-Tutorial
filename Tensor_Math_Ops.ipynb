```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4,5])
tensor_b = torch.tensor([6,7,8,9,10])
```

```python
#Addition
tensor_a + tensor_b
```

```python
#Addition Longhand
torch.add(tensor_a, tensor_b)
```

```python
#Subtraction
tensor_b - tensor_a
```

```python
#Subtraction Longhand
torch.sub(tensor_b, tensor_a)
```

```python
#Multiplication
tensor_a * tensor_b
```

```python
#Multiplication Longhand
torch.mul(tensor_a, tensor_b)
```

```python
#Division
tensor_b / tensor_a
```

```python
#Division Longhand
torch.div(tensor_b, tensor_a)
```

```python
#Modulus
tensor_b % tensor_a
```

```python
#Modulus longhand
torch.remainder(tensor_b, tensor_a)
```

```python
#Exponents
tensor_a ** tensor_b
```

```python
#Exponents Longhand
torch.pow(tensor_a, tensor_b)
```

```python
#Reassignment_
tensor_c = torch.tensor([1,2,3,4])
tensor_d = torch.tensor([5,6,7,8])
```

```python
#tensor_c = tensor_c + tensor_d (normal method)
tensor_c.add_(tensor_d) #underscore method
tensor_c
```

```python
#Another way to write longhand (Reassigned tensor_c + tensor_d)
tensor_c.add(tensor_d)
```
