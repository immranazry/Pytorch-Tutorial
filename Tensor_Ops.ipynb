```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and View (basically does the same thing)
my_torch = my_torch.reshape(2, 5)
my_torch
```

```python
# Reshape if we don't know the number of items using -1 (-1 can be used on both sides)
my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2, -1)
my_torch2
```

```python
# With reshape and view, they will update
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.reshape(2, -1)
my_torch4
```

```python
my_torch3[1] = 4141 # update value at index 1
my_torch3
```

```python
my_torch4 # it auto updates
```

```python
# Slices
my_torch5 = torch.arange(10)
my_torch5
```

```python
# Grab a specific item
my_torch5[4]
```

```python
# Grab & Slice
my_torch6 = my_torch5.reshape(5,-1)
my_torch6
```

```python
my_torch6[1]
```

```python
my_torch6[:, 1]
```

```python
# Return column
my_torch6[:, 1 :]
```
